```python
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer
```

```python
def collatz(n):
    """
    Collatz sequence
    """
    if n % 2 == 0:
        return n // 2
    else:
        return 3 * n + 1
```

```python
def collatz_sequence(n):
    """
    Collatz sequence
    """
    seq = [n]
    while n > 1:
        n = collatz(n)
        seq.append(n)
    return seq
```

```python
def random_collatz_sequence(n):
    """
    Collatz sequence
    """
    seq = [n]
    while n > 1:
        n = collatz(np.random.randint(1, n))
        seq.append(n)
    return seq
```

```python
y = np.zeros(10000, dtype=int)
x = np.arange(10000, dtype=int)

start = timer()

for i in x:
    y[i] = len(collatz_sequence(i))
fig, ax = plt.subplots()
ax.plot(x, y, ".", ms=2)

end = timer()
print(end - start)

plt.show()
```
